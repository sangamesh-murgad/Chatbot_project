```python
# Removeing unwanted characters '\xa0'

heir_ques = []
for question in heir_questions:
#     print(question.text)
    heir_ques.append(question.text)
h = []
for question in heir_ques:
    h.append(re.sub('\xa0',' ',question))
# heir_ques
h
```

As we can see from the above list, we have retrieved some of the sentences that are not questions as these sentences have the same tag as our questions.

Let us go ahead and remove these unwanted sentences from the list.

```python
#Remove first two and last four elements from the list
heir_ques = h[2:len(heir_ques) - 4]
heir_ques
```

Similarly, unwanted sentences are removed from k_means questions list too

```python
# using RE to remove unwanted characters
h = []
for x in heir_ques:
    h.append(re.sub('\d+. ',"",x).strip())
heir_ques = h
heir_ques
```

```python
k_ques = []
for question in k_questions:
    k_ques.append(question.text)
k_ques
```

```python
k_ques = k_ques[1:len(k_ques) - 4]
k_ques
```

```python
# using RE to remove unwanted characters
h = []
for x in k_ques:
    h.append(re.sub('\d+. ',"",x).strip())
k_ques = h
k_ques
```

```python
# using RE to remove unwanted characters
pca_ques = []
for question in pca_questions:
    pca_ques.append(question.text)
pca_ques
```

```python
#Third to sixth elements are deleted from pca_questions list
del pca_ques[3:7]
pca_ques
```

```python
# using RE to remove unwanted characters
h = []
for x in pca_ques:
    h.append(re.sub('\d+. ',"",x).strip())
pca_ques = h
pca_ques
```

```python
#Joining the lists
unsupervised_ques = heir_ques + k_ques + pca_ques
len(unsupervised_ques)
```

We have a total of 51 unsupervised learning questions from three different algorithms.

```python
# Creating data for EDA
EDAdata = unsupervised.copy()
EDAdata['Question_len'] = EDAdata.Questions.apply(len)
EDAdata['Q_no'] = EDAdata.index
EDAdata['Answer_len'] = EDAdata.Answers.apply(len)
```

# Length of questions in DATABASE

```python
temp = EDAdata.sort_values(by = 'Question_len',ascending = False)
print(" ")
print(f'\033[7m\033[92m\033[1m Longest Question in DATABASE is {temp.iloc[0,4]} with a length of {temp.iloc[0,3]} \033[0m',end='\n\n')
print(f'\033[1m\033[7m\033[92m Shortest Question in DATABASE is {temp.iloc[-1,4]} with a length of {temp.iloc[-1,3]} \033[0m',end='\n\n')
plt.figure(figsize=(18,8),dpi = 600)
sns.barplot(x = temp.Q_no.apply(str), y = temp.Question_len)
plt.xticks(rotation = 90,fontsize = 8)
plt.title(f'Length of each question in the DATABASE',fontsize = 16)
plt.xlabel('Question number',fontsize = 14)
plt.ylabel('Question Length',fontsize = 14)
for index,val in enumerate(temp.Question_len):
    plt.text(index-.2,val+1.5,str(val),fontsize = 9,rotation = 90)
plt.tight_layout()
plt.show()
```

# Analyzing Length of each question by Topic

```python
for label in unsupervised.Labels.unique():
    temp = EDAdata[EDAdata.Labels == label]
    temp = temp.sort_values(by = 'Question_len', ascending = False)
    print(" ")
    print(f'\033[7m\033[92m\033[1m Longest Question in {label} is {temp.iloc[0,4]} with a length of {temp.iloc[0,3]} \033[0m',end='\n\n')
    print(f'\033[1m\033[7m\033[92m Shortest Question in {label} is {temp.iloc[-1,4]} with a length of {temp.iloc[-1,3]} \033[0m',end='\n\n')
    plt.figure(figsize=(15,5))
    sns.barplot(x = temp.Q_no.apply(str), y = temp.Question_len)
    plt.xticks(rotation = 30)
    plt.title(f'Length of each question in {label}',fontsize = 16)
    plt.xlabel('Question number',fontsize = 14)
    plt.ylabel('Question Length',fontsize = 14)
    for index,val in enumerate(temp.Question_len):
        plt.text(index-.2,val+1.5,str(val),fontsize = 14)
    plt.tight_layout()
    plt.show()
```
